```python
import os
import numpy as np

from matplotlib import pyplot as plt
```

```python
file_path = "data.txt"
data = []
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        if line.strip() != '':
          data.append(int(line.strip()))
arr = np.array(data)
print(arr)
```

```python
random_copy_count_bubble = np.array([])
random_comparison_count_bubble = np.array([])
sorted_copy_count_bubble = np.array([])
sorted_comparison_count_bubble = np.array([])
reversed_copy_count_bubble = np.array([])
reversed_comparison_count_bubble = np.array([])

random_copy_count_cocktail = np.array([])
random_comparison_count_cocktail = np.array([])
sorted_copy_count_cocktail = np.array([])
sorted_comparison_count_cocktail = np.array([])
reversed_copy_count_cocktail = np.array([])
reversed_comparison_count_cocktail = np.array([])

random_copy_count_comb = np.array([])
random_comparison_count_comb = np.array([])
sorted_copy_count_comb = np.array([])
sorted_comparison_count_comb = np.array([])
reversed_copy_count_comb = np.array([])
reversed_comparison_count_comb = np.array([])
array_length = [100, 500,1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 25000, 50000]
```

```python
gap = 18
for i in range(0, len(arr), gap):
    random_copy_count_bubble = np.append(random_copy_count_bubble, arr[i])
    random_comparison_count_bubble = np.append(random_comparison_count_bubble, arr[i + 1])
    sorted_copy_count_bubble = np.append(sorted_copy_count_bubble, arr[i + 2])
    sorted_comparison_count_bubble = np.append(sorted_comparison_count_bubble, arr[i + 3])
    reversed_copy_count_bubble = np.append(reversed_copy_count_bubble, arr[i + 4])
    reversed_comparison_count_bubble = np.append(reversed_comparison_count_bubble, arr[i + 5])

    random_copy_count_cocktail = np.append(random_copy_count_cocktail, arr[i + 6])
    random_comparison_count_cocktail = np.append(random_comparison_count_cocktail, arr[i + 7])
    sorted_copy_count_cocktail = np.append(sorted_copy_count_cocktail, arr[i + 8])
    sorted_comparison_count_cocktail = np.append(sorted_comparison_count_cocktail, arr[i + 9])
    reversed_copy_count_cocktail = np.append(reversed_copy_count_cocktail, arr[i + 10])
    reversed_comparison_count_cocktail = np.append(reversed_comparison_count_cocktail, arr[i + 11])

    random_copy_count_comb = np.append(random_copy_count_comb, arr[i + 12])
    random_comparison_count_comb = np.append(random_comparison_count_comb, arr[i + 13])
    sorted_copy_count_comb = np.append(sorted_copy_count_comb, arr[i + 14])
    sorted_comparison_count_comb = np.append(sorted_comparison_count_comb, arr[i + 15])
    reversed_copy_count_comb = np.append(reversed_copy_count_comb, arr[i + 16])
    reversed_comparison_count_comb = np.append(reversed_comparison_count_comb, arr[i + 17])
print(random_comparison_count_comb)
print(random_comparison_count_bubble)
```

```python
def show_results(arr_len, bubble_arr, cocktail_arr, comb_arr,bubble_arr2, cocktail_arr2, comb_arr2 ):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 10))
    ax1.plot(arr_len, bubble_arr, color="orange", label="Bubble sort", marker='s')
    ax1.plot(arr_len, cocktail_arr, color="steelblue", label="Cocktail sort", marker='s')
    ax1.plot(arr_len, comb_arr, color="green", label="Comb sort", marker='s')
    ax1.set_xlabel("Количество элементов", fontsize=16)
    ax1.set_ylabel("Количество сравнений", fontsize=16)
    ax1.grid(True)
    ax2.plot(arr_len, bubble_arr2, color="orange", label="Bubble sort", marker='s')
    ax2.plot(arr_len, cocktail_arr2, color="steelblue", label="Cocktail sort", marker='s')
    ax2.plot(arr_len, comb_arr2, color="green", label="Comb sort", marker='s')
    plt.xlabel("Количество элементов", fontsize=16)
    plt.ylabel("Количество копирований",fontsize=16)
    ax1.legend()
    ax2.legend()
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd

sorts = ['Bubble sort','Cocktail sort', 'Comb sort']
arrays = ['Random', 'Sorted', 'Reversed']
stats = ['comp', 'copy']
sort  = [ x  for x in sorts for i in range(6) ]
array = [ x  for i in range(3) for x in arrays for i in range(2) ]
stat = [ x  for i in range(9) for x in stats  ]

df = pd.DataFrame(
              index=array_length,
              columns=[sort,array,stat])
df['Bubble sort','Random'] =  random_comparison_count_bubble
df['Bubble sort','Sorted', 'comp'] =  sorted_comparison_count_bubble
df['Bubble sort','Reversed', 'comp'] =  reversed_comparison_count_bubble

df['Cocktail sort','Random', 'comp'] =  random_comparison_count_cocktail
df['Cocktail sort','Sorted', 'comp'] =  sorted_comparison_count_cocktail
df['Cocktail sort','Reversed', 'comp'] =  reversed_comparison_count_cocktail

df['Comb sort','Random', 'comp'] =  random_comparison_count_comb
df['Comb sort','Sorted', 'comp'] =  sorted_comparison_count_comb
df['Comb sort','Reversed', 'comp'] =  reversed_comparison_count_comb

df['Bubble sort','Random', 'copy'] =  random_copy_count_bubble
df['Bubble sort','Sorted', 'copy'] =  sorted_copy_count_bubble
df['Bubble sort','Reversed', 'copy'] =  reversed_copy_count_bubble

df['Cocktail sort','Random', 'copy'] =  random_copy_count_cocktail
df['Cocktail sort','Sorted', 'copy'] =  sorted_copy_count_cocktail
df['Cocktail sort','Reversed', 'copy'] =  reversed_copy_count_cocktail

df['Comb sort','Random', 'copy'] =  random_copy_count_comb
df['Comb sort','Sorted', 'copy'] =  sorted_copy_count_comb
df['Comb sort','Reversed', 'copy'] =  reversed_copy_count_comb

print(df)
```

# **Неотсориртированный случайный массив**

```python
show_results(array_length, random_comparison_count_bubble,
             random_comparison_count_cocktail,
             random_comparison_count_comb,
             random_copy_count_bubble,
             random_copy_count_cocktail,
             random_copy_count_comb
             )
```

# **Отсортированный случайный массив**

```python
show_results(array_length, sorted_comparison_count_bubble,
             sorted_comparison_count_cocktail,
             sorted_comparison_count_comb,
             sorted_copy_count_bubble,
             sorted_copy_count_cocktail,
             sorted_copy_count_comb
             )
```

# **Обратно отсортированный случайный массив**

```python
show_results(array_length, reversed_comparison_count_bubble,
             reversed_comparison_count_cocktail,
             reversed_comparison_count_comb,
             reversed_copy_count_bubble,
             reversed_copy_count_cocktail,
             reversed_copy_count_comb
             )
```
